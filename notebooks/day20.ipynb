```python
import numpy as np
f = open("../data/data_day20.txt", "r")
data = list(f)
data = [line.strip() for line in data]
algorithm = data[0]
data = [list(line) for line in data[2:]]
data = np.array(data)
for i in range(len(data)):
    for j in range(len(data)):
        print(data[i][j], end='')
    print('')
```

```python
def get_enhancement(pos,data):
    x,y = pos
    sol = ''
    for i in [-1,0,1]:
        for j in [-1,0,1]:
            if data[y+i][x+j] == '#':
                sol+='1'
            else:
                sol+='0'
    return algorithm[int(sol,2)]

def increase_size(data):
    temp = []
    temp.append(list('.'*(len(data)+2)))
    for line in data:
        temp.append(np.concatenate((np.array(['.']),line,np.array(['.']))))
    temp.append(list('.'*(len(data)+2)))
    temp = np.array(temp)
    return temp

    
    
    
# 5776 too hight, 6772 too hight
#data = increase_size(data).copy()
#data = increase_size(data).copy()
for i in range(len(data)):
    for j in range(len(data)):
        print(data[i][j], end='')
    print('')
```
